```python
huge_json = []
for _, item in tqdm(x.iterrows()):
    if item['迁入人口比例'] > 20:
        try:
            lng1, lat1 = city_pos_dict[item['from']]
            lng2, lat2 = city_pos_dict[item['to']]
            temp = {
                'lng1': lng1,
                'lat1': lat1,
                'lng2': lng2,
                'lat2': lat2,
                'date': str(item['date']),
                'value': item['迁入人口比例']
            }
            huge_json.append(temp)
        except:
            pass
# print(huge_json)
```

```python
import json
with open('huge_json_all.json', 'w') as outfile:
    json.dump(huge_json, outfile)
```

```python
hospital_json = []
for _, city in tqdm(hospitals.iterrows()):
    for _, item in y.iterrows():
        if city['name'] in item['地名']:
            lng, lat = item['东经'], item['北纬']
            temp = {
                'lng': lng,
                'lat': lat,
                'value': city['value'],
                'name': city['name']
            }
            hospital_json.append(temp)
            break
hospital_json
```
